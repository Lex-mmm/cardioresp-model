```python


def butter_lowpass_filter(data, cutoff=0.5, fs=1000, order=6):
    """
    Apply a lowpass Butterworth filter to the given data.
    
    Parameters:
    - data: The input signal.
    - cutoff: The cutoff frequency of the filter in Hz.
    - fs: The sampling frequency of the data in Hz.
    - order: The order of the filter.
    
    Returns:
    - y: The filtered signal.
    """
    # Normalize the frequency
    nyq = 0.5 * fs
    normal_cutoff = cutoff / nyq
    
    # Get the filter coefficients
    b, a = signal.butter(order, normal_cutoff, btype='low', analog=False)
    
    # Apply the filter
    y = signal.filtfilt(b, a, data)
    return y


T=0.001;
map_10dl_1min_no=butter_lowpass_filter(ABP_10dl_1min_no, 0.15,1/T, 4);
map_10dl_1min_yes=butter_lowpass_filter(ABP_10dl_1min_yes, 0.15,1/T, 4);
map_10dl_2min_no=butter_lowpass_filter(ABP_10dl_2min_no, 0.15,1/T, 4);
map_15dl_1min_no=butter_lowpass_filter(ABP_15dl_1min_no, 0.15,1/T, 4);
map_15dl_2min_no=butter_lowpass_filter(ABP_15dl_2min_no, 0.15,1/T, 4);
map_15dl_2min_yes=butter_lowpass_filter(ABP_15dl_2min_yes, 0.15,1/T, 4);
map_20dl_1min_no=butter_lowpass_filter(ABP_20dl_1min_no, 0.15,1/T, 4);
map_20dl_2min_no=butter_lowpass_filter(ABP_20dl_2min_no, 0.15,1/T, 4);
map_20dl_2min_yes=butter_lowpass_filter(ABP_20dl_2min_yes, 0.15,1/T, 4);
map_20dl_3min_no=butter_lowpass_filter(ABP_20dl_3min_no, 0.15,1/T, 4);
map_20dl_3min_yes=butter_lowpass_filter(ABP_20dl_3min_yes, 0.15,1/T, 4);


map_10dl_1min_no_rs = signal.resample_poly(map_10dl_1min_no, up=1, down=1000)
map_10dl_1min_yes_rs = signal.resample_poly(map_10dl_1min_yes, up=1, down=1000)
map_10dl_2min_no_rs = signal.resample_poly(map_10dl_2min_no, up=1, down=1000)
map_15dl_1min_no_rs = signal.resample_poly(map_15dl_1min_no, up=1, down=1000)
map_15dl_2min_no_rs = signal.resample_poly(map_15dl_2min_no, up=1, down=1000)
map_15dl_2min_yes_rs = signal.resample_poly(map_15dl_2min_yes, up=1, down=1000)
map_20dl_1min_no_rs = signal.resample_poly(map_20dl_1min_no, up=1, down=1000)
map_20dl_2min_no_rs = signal.resample_poly(map_20dl_2min_no, up=1, down=1000)
map_20dl_2min_yes_rs = signal.resample_poly(map_20dl_2min_yes, up=1, down=1000)
map_20dl_3min_no_rs = signal.resample_poly(map_20dl_3min_no, up=1, down=1000)
map_20dl_3min_yes_rs = signal.resample_poly(map_20dl_3min_yes, up=1, down=1000)

map_10dl_1min_no_rs=map_10dl_1min_no_rs[10:-10];
map_10dl_1min_yes_rs=map_10dl_1min_yes_rs[10:-10];
map_10dl_2min_no_rs=map_10dl_2min_no_rs[10:-10];
map_15dl_1min_no_rs=map_15dl_1min_no_rs[10:-10];
map_15dl_2min_no_rs=map_15dl_2min_no_rs[10:-10];
map_15dl_2min_yes_rs=map_15dl_2min_yes_rs[10:-10];
map_20dl_1min_no_rs=map_20dl_1min_no_rs[10:-10];
map_20dl_2min_no_rs=map_20dl_2min_no_rs[10:-10];
map_20dl_2min_yes_rs=map_20dl_2min_yes_rs[10:-10];
map_20dl_3min_no_rs=map_20dl_3min_no_rs[10:-10];
map_20dl_3min_yes_rs=map_20dl_3min_yes_rs[10:-10];

plt.subplot(4, 2, 1)
plt.plot(map_10dl_1min_no_rs, 'r', label='10dl_1min_no')
plt.legend(loc='upper right')
plt.subplot(4, 2, 2)
plt.plot(map_10dl_1min_yes_rs, 'r', label='10dl_1min_yes')
plt.legend(loc='upper right')
plt.subplot(4, 2, 3)
plt.plot(map_10dl_2min_no_rs, 'r', label='10dl_2min_no')
plt.legend(loc='upper right')
plt.subplot(4, 2, 4)
plt.plot(map_15dl_1min_no_rs, 'r', label='15dl_1min_no')
plt.legend(loc='upper right')
plt.subplot(4, 2, 5)
plt.plot(map_15dl_2min_no_rs, 'r', label='15dl_2min_no')
plt.legend(loc='upper right')
plt.subplot(4, 2, 6)
plt.plot(map_15dl_2min_yes_rs, 'r', label='15dl_2min_yes')
plt.legend(loc='upper right')
plt.subplot(4, 2, 7)
plt.plot(map_20dl_3min_yes_rs, 'r', label='20dl_3min_yes')
plt.legend(loc='upper right')
plt.subplot(4, 2, 8)
plt.plot(map_20dl_3min_no_rs, 'r', label='20dl_3min_no')
plt.legend(loc='upper right')



# merge the data into one array
data = np.array([map_10dl_1min_no_rs, map_10dl_1min_yes_rs, map_10dl_2min_no_rs, map_15dl_1min_no_rs, map_15dl_2min_no_rs, map_15dl_2min_yes_rs, map_20dl_1min_no_rs, map_20dl_2min_no_rs, map_20dl_2min_yes_rs, map_20dl_3min_no_rs, map_20dl_3min_yes_rs])
# add header to the data
header = "map_10dl_1min_no_rs, map_10dl_1min_yes_rs, map_10dl_2min_no_rs, map_15dl_1min_no_rs, map_15dl_2min_no_rs, map_15dl_2min_yes_rs, map_20dl_1min_no_rs, map_20dl_2min_no_rs, map_20dl_2min_yes_rs, map_20dl_3min_no_rs, map_20dl_3min_yes_rs"
# save the data to a file
np.savetxt('/Users/l.m.vanloon/Library/CloudStorage/OneDrive-UMCUtrecht/Modelling /cardioresp model/code/outputs/ABP_resampled.csv', data.T, delimiter=',', header=header, comments='')
```
